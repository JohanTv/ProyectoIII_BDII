```python
import pandas as pd
import pickle
# dataset_path = "./data/datasetv2.csv"
# scaler_path = "./bin/scaler.dat"
# pca_path = "./bin/pca.dat"
# ncomponents_path = "./bin/ncomponents.dat"
dataset_path = "./data/dataset_12800.csv"
scaler_path = "./bin/scaler.dat"
pca_path = "./bin/pca.dat"
ncomponents_path = "./bin/ncomponents.dat"

df = pd.read_csv(dataset_path)
scaler = pickle.load(open(scaler_path, "rb"))
pca = pickle.load(open(pca_path, "rb"))
ncomponents = pickle.load(open(ncomponents_path, "rb"))

print("Completed")
df.tail(5)
```

```python
features = [str(i) for i in range(1, ncomponents+1)]
x = df.loc[:, features]
y = df.loc[:, ["path"]]
y.tail(5)
```

```python
import numpy as np
from rtree import index
nrows = df.shape[0]

path = "./bin_test/rtree_index"
p = index.Property()
p.dimension = ncomponents #D

idx = index.Index(path, properties=p)

for i in range(nrows):
    v = x.iloc[i].values
    image = y.iloc[i].values[0]
    point = tuple(np.concatenate([v, v]))
    idx.insert(i, point, obj=image)

idx.close()
print("rtree ready")
```
